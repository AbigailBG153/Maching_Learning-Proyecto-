#Recoleccion de criptomonedas por categoria

```python
import requests
import pandas as pd
import time

# Función para realizar solicitudes a la API de CoinMarketCap
def get_data(url, offset=None):
    headers = {
        'X-CMC_PRO_API_KEY': '830a9b54-7134-4fb9-b943-e3641f251859',  # Reemplaza 'tu_api_key' con tu propia clave de API
        'Accept': 'application/json'
    }
    params = {'start': offset} if offset is not None else None  # Agregamos el parámetro de offset si está presente
    response = requests.get(url, headers=headers, params=params)
    return response.json()

# Función para obtener el ID de una categoría dado su nombre
def get_category_id(category_name):
    categories_url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/categories'
    categories_data = get_data(categories_url)
    if 'data' in categories_data:
        for category in categories_data['data']:
            if category['name'].lower() == category_name.lower():
                print(f"ID de la categoría '{category_name}': {category['id']}")
                return category['id']
    else:
        print("Error: No se pudo obtener la lista de categorías.")
    return None

# Función para obtener las monedas de una categoría en grupos de 100
def get_coins_by_category(category_id, offset):
    coins_url = f'https://pro-api.coinmarketcap.com/v1/cryptocurrency/category?id={category_id}'
    coins_data = get_data(coins_url, offset)['data']['coins']
    return coins_data

# Lista de nombres de categorías
category_names = ['Gaming', 'Memes', 'AI & Big Data']  # Ajustar según las categorías deseadas

# DataFrame para almacenar la información de las monedas
coins_data = []

# Diccionario para almacenar la cantidad de monedas encontradas por categoría
coins_count = {}

# Iterar sobre cada nombre de categoría
for category_name in category_names:
    category_id = get_category_id(category_name)
    if category_id:
        print(f"Analizando monedas en la categoría: '{category_name}'")

        # Inicializar el offset en 1
        offset = 1

        # Realizar consultas de las monedas hasta que ya no haya más monedas en la categoría
        while True:
            coins = get_coins_by_category(category_id, offset)

            # Si no hay monedas en la respuesta, salimos del bucle
            if not coins:
                break

            # Guardar las monedas encontradas en la lista de datos
            coins_data.extend([{'Category ID': category_id, 'Category Name': category_name, 'Coins': coin} for coin in coins])

            # Incrementar el offset para la próxima consulta
            offset += len(coins)

            # Imprimir las monedas encontradas
            print(f"Monedas encontradas en la categoría '{category_name}': {[coin['name'] for coin in coins]}")

            # Esperar un minuto antes de realizar la siguiente consulta
            time.sleep(60)

        # Contar la cantidad de monedas encontradas
        num_coins = len(coins_data)
        coins_count[category_name] = num_coins

        # Imprimir detalle del análisis
        print(f"Análisis completo de la categoría '{category_name}': Se encontraron {num_coins} monedas.")

# Calcular el total de monedas encontradas en todas las categorías
total_coins = sum(coins_count.values())

# Imprimir el total de monedas encontradas
print(f"Total de monedas encontradas en todas las categorías: {total_coins}")

# Convertir la lista de datos en un DataFrame
coins_df = pd.DataFrame(coins_data)

# Guardar el DataFrame en un archivo CSV
coins_df.to_csv('coins_data3.csv', index=False)

# Imprimir el DataFrame
print(coins_df)

print("Programa finalizado.")
```

```python
import ast
import pandas as pd

# Crear una lista vacía para almacenar los datos
data_list = []

# Supongamos que data_categoria['Coins'] es una lista de cadenas que representan diccionarios
for cadena_diccionario, category_name in zip(data_categoria['Coins'], data_categoria['Category Name']):
    # Convertir la cadena en un diccionario utilizando ast.literal_eval
    coin_dict = ast.literal_eval(cadena_diccionario)

    # Agregar 'Category Name' al diccionario
    coin_dict['Category Name'] = category_name

    # Agregar el diccionario a la lista
    data_list.append(coin_dict)

# Convertir la lista de diccionarios en un DataFrame
df = pd.DataFrame(data_list)

# Imprimir el DataFrame resultante
df
```

```python
import requests
import pandas as pd

# URL de la API de CoinGecko para obtener información de los exchanges
url = 'https://api.coingecko.com/api/v3/exchanges'

# Realizar la solicitud GET a la API
response = requests.get(url)

# Verificar si la solicitud fue exitosa (código de estado 200)
if response.status_code == 200:
    # Convertir los datos de respuesta JSON a un DataFrame de pandas
    data = response.json()
    df = pd.DataFrame(data)

    # Guardar el DataFrame en un archivo CSV
    df.to_csv('exchanges.csv', index=False)

    print("Los datos se han guardado correctamente en exchanges.csv.")
else:
    print("Error al obtener los datos de la API.")
```
