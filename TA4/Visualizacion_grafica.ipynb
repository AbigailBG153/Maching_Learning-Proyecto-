```python
import pandas as pd
import matplotlib.pyplot as plt


# 1. Gráfico de Barras Horizontal - Comparación del Número Total de Tokens por Portafolio
plt.figure(figsize=(10, 6))
bars = plt.barh(datos_portafolio['name'], datos_portafolio['num_tokens'], color='skyblue')

# Diseño minimalista
plt.xlabel('Número Total de Tokens', fontsize=12)
plt.ylabel('Nombre del Portafolio', fontsize=12)
plt.title('Número Total de Tokens por Portafolio', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Agregar los valores en la parte derecha de cada barra
for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height()/2, f'{int(width)}', va='center', ha='left', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv('/content/data_portfologio_token.csv',  delimiter=';')
data.drop(data.columns[0], axis=1, inplace=True)
data = data.iloc[:, :-1]
data = data.drop_duplicates(subset=['portfolio', 'nombre_token', 'symbol_token', 'categoria'])
data.reset_index(drop=True, inplace=True)
data
```

```python
data.at[351, 'categoria'] = 'Meme'
```

```python
data.info()
```

```python
import matplotlib.pyplot as plt

# Contar la cantidad de tokens en cada categoría
category_counts = data['categoria'].value_counts()

# Gráfico de Barras
plt.figure(figsize=(10, 6))
bars = plt.bar(category_counts.index, category_counts.values, color='skyblue')
plt.xlabel('Categoría')
plt.ylabel('Número de Tokens')
plt.title('Número de tokens en cada categoría')

# Agregar los valores en la parte superior de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom', ha='center')

plt.tight_layout()
plt.show()

# Gráfico de Pastel
plt.figure(figsize=(10, 6))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'orange'], startangle=140)
plt.title('Distribución de tokens por categoría')
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Agrupar por portfolio y categoría y contar la cantidad de tokens
conteo_por_categoria = data.groupby(['portfolio', 'categoria']).size().unstack(fill_value=0)

# Obtener los nombres de los portfolios
portfolios = conteo_por_categoria.index

# Obtener las categorías
categorias = conteo_por_categoria.columns

# Crear el gráfico de barras
plt.figure(figsize=(12, 8))

# Definir colores para las categorías
colores = ['blue', 'green', 'red', 'orange']

# Iterar sobre los portfolios
for i, portfolio in enumerate(portfolios):
    # Obtener las cantidades de tokens por categoría para el portfolio actual
    cantidades_por_categoria = conteo_por_categoria.loc[portfolio]

    # Crear una lista de colores para las barras dentro de cada portfolio
    colores_barras = [colores[categorias.tolist().index(c)] for c in cantidades_por_categoria.index]

    # Crear la barra apilada para el portfolio actual
    plt.barh(portfolio, cantidades_por_categoria, color=colores_barras, label=categorias if i == 0 else None)

# Personalizar el gráfico
plt.xlabel('Cantidad de Tokens')
plt.ylabel('Portfolio')
plt.title('Cantidad de Tokens por Portfolio y Categoría')
plt.legend(title='Categorías')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supongamos que 'data' es el DataFrame que contiene tu información

# Contar la cantidad de veces que aparece cada token en todos los portafolios
conteo_tokens = data['nombre_token'].value_counts()

# Filtrar los 10 tokens más repetidos
top_tokens = conteo_tokens.head(10)

# Definir colores para cada categoría
colores_categorias = {'IA': 'blue', 'Gaming': 'green', 'RWA': 'red', 'Meme': 'orange'}

# Crear el gráfico de barras horizontal con colores específicos para cada categoría
plt.figure(figsize=(10, 6))
bar_colors = [colores_categorias[data[data['nombre_token'] == token]['categoria'].values[0]] for token in top_tokens.index]
bars = plt.barh(top_tokens.index, top_tokens.values, color=bar_colors)
plt.xlabel('Cantidad de Veces Repetido')
plt.ylabel('Token')
plt.title('Tokens Más Repetidos en los Portafolios')

# Agregar leyenda
legend_labels = {v: k for k, v in colores_categorias.items()}  # Invertir el diccionario para usarlo en la leyenda
handles = [plt.Rectangle((0, 0), 1, 1, color=color) for color in colores_categorias.values()]
plt.legend(handles, legend_labels.values(), title='Categoría')

plt.tight_layout()
plt.show()
```
